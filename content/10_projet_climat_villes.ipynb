## Correction 1

```python
df["mois"] = pd.to_datetime(df["mois"])
print(df["mois"].dtype)
```

## Correction 2

```python
m = df["mois"].dt.month

conditions = [
    m.isin([1, 2]),
    m.isin([3, 4, 5]),
    m.isin([6])
]
choix = ["hiver", "printemps", "ete"]

df["saison"] = np.select(conditions, choix, default="autre")
df[["mois", "saison"]].head()
```

## Correction 3

```python
df["indice_confort"] = df["tavg"] - 0.2 * df["wind_kmh"] + 0.05 * df["humidite"]
df[["ville", "mois", "indice_confort"]]
```

## Correction 4

```python
synthese_ville = df.groupby("ville").agg(
    temperature_moyenne=("tavg", "mean"),
    precipitations_totales=("precip_mm", "sum"),
    vent_moyen=("wind_kmh", "mean")
)
synthese_ville
```

## Correction 5

```python
df[(df["tavg"] >= 15) & (df["precip_mm"] < 50)]
```

## Correction 6

```python
df.pivot_table(index="mois", columns="ville", values="tavg", aggfunc="mean")
```

## Correction 7

```python
df.isna().sum()

df_clean = df.copy()
df_clean["precip_mm"] = df_clean["precip_mm"].fillna(df_clean["precip_mm"].mean())
df_clean
```

## Correction 8

```python
sns.lineplot(data=df, x="mois", y="tavg", hue="ville", marker="o")
plt.title("Évolution des températures moyennes")
plt.show()
```

## Correction 9

```python
precip_ville = df.groupby("ville", as_index=False)["precip_mm"].sum()

sns.barplot(data=precip_ville, x="ville", y="precip_mm")
plt.title("Précipitations totales par ville")
plt.show()
```

## Correction 10

```python
synthese_ville = df.groupby("ville").agg(
    temperature_moyenne=("tavg", "mean"),
    precipitations_totales=("precip_mm", "sum")
)

ville_plus_chaude = synthese_ville["temperature_moyenne"].idxmax()
ville_plus_humide = synthese_ville["precipitations_totales"].idxmax()

print("Ville la plus chaude :", ville_plus_chaude)
print("Ville la plus humide :", ville_plus_humide)
```

## Correction 11

```python
df["indice_confort"] = df["tavg"] - 0.2 * df["wind_kmh"] + 0.05 * df["humidite"]

tableau_final = (
    df.groupby("ville")
    .agg(
        temperature_moyenne=("tavg", "mean"),
        precipitations_totales=("precip_mm", "sum"),
        vent_moyen=("wind_kmh", "mean"),
        indice_confort_moyen=("indice_confort", "mean")
    )
    .sort_values("temperature_moyenne", ascending=False)
)

tableau_final
```
