## Correction 2

```python
df["indice_chaleur"] = df["tavg_annuelle"] + 0.05 * df["jours_chauds"]
df[["ville", "indice_chaleur"]]
```

## Correction 3

```python
df.sort_values("tavg_annuelle", ascending=False)
```

## Correction 4

```python
df.groupby("zone")["tavg_annuelle"].mean()
```

## Correction 5

```python
df[(df["tavg_annuelle"] >= 13) & (df["precip_annuelle"] < 850)]
```

## Correction 6

```python
plt.figure(figsize=(7, 6))
plt.scatter(df["longitude"], df["latitude"])

for _, row in df.iterrows():
    plt.annotate(row["ville"], (row["longitude"], row["latitude"]))

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Localisation des villes")
plt.show()
```

## Correction 7

```python
plt.figure(figsize=(7, 6))
sc = plt.scatter(
    df["longitude"],
    df["latitude"],
    s=df["jours_chauds"] * 8,
    c=df["tavg_annuelle"]
)

for _, row in df.iterrows():
    plt.annotate(row["ville"], (row["longitude"], row["latitude"]))

plt.colorbar(sc, label="Température annuelle moyenne")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Carte climatique simplifiée")
plt.show()
```

## Correction 8

```python
sns.barplot(data=df, x="ville", y="precip_annuelle")
plt.title("Précipitations annuelles par ville")
plt.xticks(rotation=20)
plt.show()
```

## Correction 9

```python
synthese_zone = df.groupby("zone").agg(
    temperature_moyenne=("tavg_annuelle", "mean"),
    precipitations_moyennes=("precip_annuelle", "mean"),
    jours_chauds_moyens=("jours_chauds", "mean")
)
synthese_zone
```

## Correction 10

```python
print("Ville la plus chaude :", df.loc[df["tavg_annuelle"].idxmax(), "ville"])
print("Ville la plus pluvieuse :", df.loc[df["precip_annuelle"].idxmax(), "ville"])
print("Ville avec le plus de jours chauds :", df.loc[df["jours_chauds"].idxmax(), "ville"])
```
