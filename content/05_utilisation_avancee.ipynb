## Correction du mini défi

```python
class StationAnalyse:
    def __init__(self, nom, temperatures):
        self.nom = nom
        self.temperatures = temperatures

    @property
    def moyenne(self):
        return sum(self.temperatures) / len(self.temperatures)

    def amplitude(self):
        return max(self.temperatures) - min(self.temperatures)

    def resume(self):
        return (
            f"Station {self.nom} : moyenne = {self.moyenne:.1f} °C, "
            f"amplitude = {self.amplitude():.1f} °C"
        )

s1 = StationAnalyse("Paris", [6.0, 8.5, 12.0, 14.0])
s2 = StationAnalyse("Marseille", [9.0, 11.5, 15.0, 18.0])

print(s1.resume())
print(s2.resume())
```
