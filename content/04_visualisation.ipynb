## Correction 1

```python
paris = df[df["ville"] == "Paris"]

plt.plot(paris["mois"], paris["tavg"])
plt.title("Température moyenne à Paris")
plt.xlabel("Mois")
plt.ylabel("Température moyenne (°C)")
plt.show()
```

## Correction 2

```python
plt.hist(df["precip_mm"], bins=6)
plt.title("Histogramme des précipitations")
plt.show()
```

## Correction 3

```python
plt.scatter(df["humidite"], df["tavg"])
plt.xlabel("Humidité (%)")
plt.ylabel("Température moyenne (°C)")
plt.title("Humidité vs température")
plt.show()
```

## Correction 4

```python
sns.barplot(data=df, x="ville", y="tavg")
plt.title("Température moyenne par ville")
plt.show()
```

## Correction 5

```python
sns.boxplot(data=df, x="ville", y="precip_mm")
plt.title("Précipitations par ville")
plt.show()
```

## Correction 6

```python
sns.histplot(data=df, x="tavg")
plt.title("Distribution des températures moyennes")
plt.show()
```

## Correction 7

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x="ville", y="tavg")
plt.xticks(rotation=20)
plt.title("Température moyenne par ville")
plt.show()
```

## Correction 8

```python
sns.lineplot(data=df, x="mois", y="precip_mm", hue="ville", marker="o")
plt.title("Évolution des précipitations")
plt.show()
```

## Correction 9

```python
sns.scatterplot(data=df, x="humidite", y="tavg", hue="ville")
plt.title("Humidité vs température")
plt.show()
```

## Correction 10

```python
sns.regplot(data=df, x="wind_kmh", y="precip_mm")
plt.title("Vent vs précipitations")
plt.show()
```

## Correction 11

```python
df_precip = df.groupby("ville", as_index=False)["precip_mm"].sum()

sns.barplot(data=df_precip, x="ville", y="precip_mm")
plt.title("Somme des précipitations par ville")
plt.show()
```

## Correction 12

```python
sns.countplot(data=df, x="ville")
plt.title("Nombre d'observations par ville")
plt.show()
```

## Correction 13

```python
sns.histplot(data=df, x="wind_kmh", kde=True)
plt.title("Distribution du vent")
plt.show()
```

## Correction 14

```python
sns.violinplot(data=df, x="ville", y="tavg")
plt.title("Distribution des températures par ville")
plt.show()
```

## Correction 15

```python
corr = df[["tavg", "precip_mm", "wind_kmh", "humidite"]].corr()

sns.heatmap(corr, annot=True)
plt.title("Matrice de corrélation")
plt.show()
```

## Correction 16

```python
sns.barplot(data=df, x="ville", y="tavg")
plt.title("Température moyenne par ville")
plt.savefig("temperature_par_ville.png")
plt.show()
```

## Correction du mini défi

```python
# 1. Comparaison des températures moyennes par ville
sns.barplot(data=df, x="ville", y="tavg")
plt.title("Température moyenne par ville")
plt.show()

# 2. Distribution des précipitations
sns.histplot(data=df, x="precip_mm", kde=True)
plt.title("Distribution des précipitations")
plt.show()

# 3. Relation entre humidité et température
sns.scatterplot(data=df, x="humidite", y="tavg", hue="ville")
plt.title("Humidité vs température")
plt.show()
```
