## Correction du mini défi

````python
temperatures = np.random.randint(0, 35, 20)
print("Températures :", temperatures)

temp_chaudes = temperatures[temperatures >= 20]
print("Températures >= 20 :", temp_chaudes)

if len(temp_chaudes) > 0:
    print("Moyenne des températures chaudes :", np.mean(temp_chaudes))
else:
    print("Aucune température chaude")

etat = np.where(temperatures >= 20, "chaud", "frais")
print("Statut :", etat)
````
