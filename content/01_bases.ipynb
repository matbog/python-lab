## 9. Exceptions

### Exercice 20
Teste la division `10 / 0` dans un bloc `try / except` et affiche un message clair si une erreur se produit.

```python
# À compléter

try:
    resultat = 10 / 0
    print(resultat)
except ZeroDivisionError:
    print("Erreur : division par zéro")
```

### Exercice 21

```python
# À compléter

try:
    nombre = int("abc")
    print(nombre)
except ValueError:
    print("Erreur : conversion impossible en entier")
```

## 11. Mini défi final

Crée un petit programme qui :

1. stocke une liste de notes
2. calcule la moyenne
3. affiche `"Admis"` si la moyenne est au moins égale à 10
4. sinon affiche `"Ajourné"`

Essaie d'écrire la solution toi-même avant de regarder un corrigé.

```python
# À compléter

notes = [12, 9, 14, 11]
moyenne = sum(notes) / len(notes)

print("Moyenne :", moyenne)

if moyenne >= 10:
    print("Admis")
else:
    print("Ajourné")
```
