# Comment utiliser ce notebook

## ⚙️ Exécuter du code

1. Cliquer dans une cellule
2. Appuyer sur **Shift + Enter**

▶️ Essaie ci-dessous : 

```python
print("Hello World")
```

### ✏️ Modifier le code

On peut modifier n'importe quelle cellule.

▶️ Exemple :

```python
print("Modifie ce texte puis exécute la cellule")
```

### ➕ Ajouter une cellule

- Bouton "+" à droite

### ❌ En cas d'erreur

- Lire le message affiché sous la cellule.

▶️ Exemple :

```python
print(x)
```

▶️ Ici, `x` n'existe pas → erreur normale
