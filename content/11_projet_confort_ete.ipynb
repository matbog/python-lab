## Correction 1

```python
df["jour_chaud"] = np.where(df["tmax"] >= 30, 1, 0)
df[["tmax", "jour_chaud"]].head()
```

## Correction 2

```python
df["jour_tres_chaud"] = np.where(df["tmax"] >= 35, 1, 0)
df[["tmax", "jour_tres_chaud"]].head()
```

## Correction 3

```python
df["nuit_chaude"] = np.where(df["tmin"] >= 20, 1, 0)
df[["tmin", "nuit_chaude"]].head()
```

## Correction 4

```python
synthese = df.groupby("ville").agg(
    nb_jours_chauds=("jour_chaud", "sum"),
    nb_jours_tres_chauds=("jour_tres_chaud", "sum"),
    nb_nuits_chaudes=("nuit_chaude", "sum"),
    temperature_moyenne=("tavg", "mean")
)
synthese
```

## Correction 5

```python
df[df["tmax"] >= 35]
```

## Correction 6

```python
conditions = [
    df["tmax"] < 30,
    (df["tmax"] >= 30) & (df["tmax"] < 35),
    df["tmax"] >= 35
]
choix = ["modere", "fort", "extreme"]

df["classe_chaleur"] = np.select(conditions, choix)
df[["tmax", "classe_chaleur"]].head()
```

## Correction 7

```python
sns.lineplot(data=df, x="date", y="tmax", hue="ville", marker="o")
plt.title("Évolution des températures maximales")
plt.show()
```

## Correction 8

```python
jours_chauds = df.groupby("ville", as_index=False)["jour_chaud"].sum()

sns.barplot(data=jours_chauds, x="ville", y="jour_chaud")
plt.title("Nombre de jours chauds")
plt.show()
```

## Correction 9

```python
synthese = df.groupby("ville").agg(
    nb_jours_tres_chauds=("jour_tres_chaud", "sum"),
    temperature_moyenne=("tavg", "mean")
)

print("Ville la plus exposée aux jours très chauds :", synthese["nb_jours_tres_chauds"].idxmax())
print("Ville la plus chaude en moyenne :", synthese["temperature_moyenne"].idxmax())
```

## Correction 10

```python
tableau_final = (
    df.groupby("ville")
    .agg(
        nb_jours_chauds=("jour_chaud", "sum"),
        nb_jours_tres_chauds=("jour_tres_chaud", "sum"),
        nb_nuits_chaudes=("nuit_chaude", "sum"),
        temperature_moyenne=("tavg", "mean")
    )
    .sort_values(
        ["nb_jours_tres_chauds", "nb_nuits_chaudes", "temperature_moyenne"],
        ascending=False
    )
)

tableau_final
```
