## Correction 1

```python
class SerieClimatique:
    def __init__(self, ville, temperatures):
        self.ville = ville
        self.temperatures = np.array(temperatures)

    @property
    def moyenne(self):
        return np.mean(self.temperatures)

    def amplitude(self):
        return np.max(self.temperatures) - np.min(self.temperatures)

    def resume(self):
        return (
            f"{self.ville} : moyenne = {self.moyenne:.1f} °C, "
            f"amplitude = {self.amplitude():.1f} °C"
        )
```

```python
paris = SerieClimatique("Paris", donnees["Paris"])
marseille = SerieClimatique("Marseille", donnees["Marseille"])

print(paris.resume())
print(marseille.resume())
```

## Correction 3

```python
class SerieClimatique:
    def __init__(self, ville, temperatures):
        self.ville = ville
        self.temperatures = np.array(temperatures)

    @property
    def moyenne(self):
        return np.mean(self.temperatures)

    def amplitude(self):
        return np.max(self.temperatures) - np.min(self.temperatures)

class SerieClimatiqueComparee(SerieClimatique):
    def nb_mois_chauds(self, seuil=20):
        return np.sum(self.temperatures >= seuil)
```

## Correction 4

```python
lyon = SerieClimatiqueComparee("Lyon", donnees["Lyon"])
print(lyon.nb_mois_chauds())
```

## Correction 11

```python
arr = np.random.uniform(0, 35, 200000)
liste = arr.tolist()

t0 = time.perf_counter()
total = 0
for x in liste:
    total += x
moy_python = total / len(liste)
t1 = time.perf_counter()

t2 = time.perf_counter()
moy_numpy = np.mean(arr)
t3 = time.perf_counter()

print("Moyenne Python :", moy_python)
print("Moyenne NumPy  :", moy_numpy)
print("Temps Python   :", t1 - t0)
print("Temps NumPy    :", t3 - t2)
```

## Correction 12

```python
arr = np.random.uniform(0, 35, 200000)
liste = arr.tolist()

t0 = time.perf_counter()
compteur = 0
for x in liste:
    if x >= 20:
        compteur += 1
t1 = time.perf_counter()

t2 = time.perf_counter()
compteur_numpy = np.sum(arr >= 20)
t3 = time.perf_counter()

print("Count Python :", compteur)
print("Count NumPy  :", compteur_numpy)
print("Temps Python :", t1 - t0)
print("Temps NumPy  :", t3 - t2)
```

## Correction 13

```python
synthese = df_long.groupby("ville").agg(
    temperature_moyenne=("temperature", "mean"),
    temperature_max=("temperature", "max"),
    temperature_min=("temperature", "min")
)
synthese["amplitude"] = synthese["temperature_max"] - synthese["temperature_min"]
synthese["nb_mois_chauds"] = df_long.groupby("ville")["temperature"].apply(lambda s: np.sum(s >= 20))

print("Ville la plus chaude :", synthese["temperature_moyenne"].idxmax())
print("Ville avec la plus forte amplitude :", synthese["amplitude"].idxmax())
print("Ville avec le plus de mois chauds :", synthese["nb_mois_chauds"].idxmax())
```

## Correction 14

```python
tableau_final = df_long.groupby("ville").agg(
    temperature_moyenne=("temperature", "mean"),
    temperature_max=("temperature", "max"),
    temperature_min=("temperature", "min"),
    nb_mois_chauds=("temperature", lambda s: np.sum(s >= 20))
)

tableau_final["amplitude"] = tableau_final["temperature_max"] - tableau_final["temperature_min"]

tableau_final = tableau_final[
    ["temperature_moyenne", "temperature_max", "temperature_min", "amplitude", "nb_mois_chauds"]
].sort_values("temperature_moyenne", ascending=False)

tableau_final
```
